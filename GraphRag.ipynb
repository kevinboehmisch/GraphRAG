1.Vektor-Suche (Das "Bauchgefühl")
Vektor-Suche (klassisches RAG) sucht nach semantischer Ähnlichkeit. Das heißt: "Klingt der Inhalt ähnlich?"

Toy Story Plot: "Spielzeuge werden lebendig, wenn Menschen nicht hinsehen."

Jumanji Plot: "Ein magisches Brettspiel lässt Dschungeltiere lebendig werden und Gefahr droht."

Was macht der Vektor? Er sieht in beiden: Dinge werden lebendig, Kinder, Abenteuer, Magie. Deshalb würde eine Vektor-Suche Jumanji tatsächlich wahrscheinlich finden, wenn du nach "Filmen mit lebendigen Objekten" suchst.

Aber wo versagt der Vektor? Wenn du fragst: "Zeige mir Filme vom gleichen Regisseur wie Toy Story."

Der Vektor sucht im Plot von Toy Story. Da steht vielleicht nichts über den Regisseur.

Er findet dann vielleicht Small Soldiers (weil es da auch um lebendige Spielzeuge geht), obwohl das ein ganz anderer Regisseur ist.

Ergebnis: Semantisch ähnlich (Thema), aber faktisch falsch (Regisseur).

GraphRAG (Das "Wissen")GraphRAG kombiniert das Bauchgefühl mit harten Fakten.Hier glänzt GraphRAG bei Fällen, die semantisch komplett unterschiedlich sind, aber logisch zusammengehören.Beispiel: Toy Story $\leftrightarrow$ Cast Away (Verschollen)Plot Toy Story: Bunter Kinderfilm, Plastikspielzeug.Plot Cast Away: Mann strandet einsam auf Insel, Überlebenskampf, redet mit Volleyball.Vektor-Check:Die Vektoren liegen meilenweit auseinander. "Kinderzimmer" vs. "Einsame Insel". Der Vektor würde Cast Away niemals vorschlagen, wenn du über Toy Story redest.GraphRAG-Check:Der Graph kennt die Struktur:(Toy Story)-[:ACTED_IN]-(Tom Hanks)-[:ACTED_IN]-(Cast Away)Wenn du GraphRAG fragst: "Was hat Toy Story mit Cast Away zu tun?"Der Graph springt von Toy Story zu Tom Hanks.Von Tom Hanks springt er zu Cast Away.Die KI antwortet: "Obwohl die Filme thematisch anders sind, spielt in beiden Tom Hanks die Hauptrolle (in Toy Story als Stimme von Woody)."

Vektor Vorteile und Nachteile:

Vectors work well for:

Contextual or Meaning Based Questions

Fuzzy or Vague queries

Broad or Open-Ended questions

Complex queries with multiple concepts

Vectors are ineffective for:

Highly Specific or Fact-Based Questions

Numerical or Exact-Match Queries

Boolean or Logical Queries

Ambiguous or Unclear Queries without Context

Specialized Knowledge

GraphRAG Vorteile:

Benefits of GraphRAG:

Richer Context

Graphs capture relationships between entities, enabling retrieval of more relevant and connected information.

Improved Accuracy

By combining vector similarity with graph traversal, results are more precise and context-aware.

Explainability

Graphs provide clear paths and connections, making it easier to understand why certain results were retrieved.

Flexible Queries:

GraphRAG supports complex queries, such as combining full-text, vector, and text-to-cypher searches.

Enhanced Reasoning

Graphs enable reasoning over data, supporting advanced use cases like recommendations and knowledge discovery.

```python
from neo4j import GraphDatabase

# Deine Daten (Passwort hast du eben im Browser geändert!)
URI = "bolt://localhost:7687"
AUTH = ("neo4j", "password") 

def run_cypher(query):
    # Verbindung aufbauen
    with GraphDatabase.driver(URI, auth=AUTH) as driver:
        driver.verify_connectivity()
        # Query ausführen
        records, summary = driver.execute_query(query)
        return records

# TEST: Wir schreiben was rein
run_cypher("CREATE (n:TestNode {text: 'Hallo Jupyter!'})")

# TEST: Wir lesen es aus
result = run_cypher("MATCH (n:TestNode) RETURN n.text AS text")

for record in result:
    print(record["text"])
```
